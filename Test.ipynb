```python
# Metadata about journal articles in CORD-19 Dataset
bibs = pd.read_csv("bibs.csv")
bibs
```

```python
# Clinical data on patients tested for covid-19
diagnosis = pd.read_csv("diag wide.csv")
diagnosis
```

Purpose is to extrapolate data from various sources to output vaccines or therapies for a given symptom or DNA sequence

```python
def search_term(word, df=bibs):
    bi = []
    for i in range(len(df["title"])):
        try:
            if word.lower() in df["title"][i].lower():
                bi.append(df["url"][i])
        except:
            pass
    return bi
```

```python
def search_diagnosis(word, df=diagnosis):
    diag = []
    for i in range(len(df)):
        try:
            if word.lower() in df['clinicalNotes'][i].lower():
                diag.append(df['testResults'][i])
        except:
            pass
    return diag            
```

```python
word = "pneumonia"
search_term(word)
```

```python
search_diagnosis(word)
```
