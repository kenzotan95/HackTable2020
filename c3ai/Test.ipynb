```python
def prob_covid(df):
    gender = []
    
    for i in range(len(df)):
        try:
            if 'woman' in df['clinicalNotes'][i].lower() or 'female' in df['clinicalNotes'][i].lower():
                gender.append("Female")
            elif 'man' in df['clinicalNotes'][i].lower() or 'male' in df['clinicalNotes'][i].lower():
                gender.append("Male")
            else:
                gender.append("N/A")
        except:
            pass
        
    df['Gender'] = gender
    
```
